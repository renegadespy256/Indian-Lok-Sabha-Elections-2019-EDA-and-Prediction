```python
df1 = data[['STATE','CONSTITUENCY','WINNER','Party New','SYMBOL','GENDER','CRIMINAL\nCASES','AGE','CATEGORY','EDUCATION','TOTAL\nVOTES','TOTAL ELECTORS','ASSETS','LIABILITIES']]
cat_cols = ['STATE','CONSTITUENCY','Party New','SYMBOL','GENDER','CATEGORY','EDUCATION','ASSETS','LIABILITIES']
num_cols = ['CRIMINAL\nCASES','AGE','TOTAL\nVOTES','TOTAL ELECTORS']
```

```python
df_winner = data['WINNER'].value_counts().reset_index()
df_winner.columns = ['RESULT','COUNT']
pie = go.Pie(labels=df_winner['RESULT'], values=df_winner['COUNT'], marker=dict(line=dict(color='black', width=1)))
layout = go.Layout(title='Total Candidates vs Winners')
fig = go.Figure(data=[pie], layout=layout)
py.iplot(fig)
```

```python
dataset = pd.get_dummies(df1, columns = cat_cols)
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
columns_to_scale = num_cols
dataset[columns_to_scale] = Scaler.fit_transform(dataset[columns_to_scale])
dataset.head()
```

```python
dataset.describe()
```

```python
from sklearn.utils import resample
df_majority = dataset[dataset.WINNER == 0]
df_minority = dataset[dataset.WINNER == 1]
df_minority_upsampled = resample(df_minority, replace = True,n_samples = 1452, random_state = 0) 
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
df_upsampled.WINNER.value_counts()
```

Dataset is balanced.

```python
y = df_upsampled['WINNER']
X = df_upsampled.drop(['WINNER'], axis = 1)
```

```python
from xgboost import XGBClassifier
```

```python

from sklearn.model_selection import cross_val_score
rfc_scores = []
for k in range(1,21):
    classifier= XGBClassifier(n_estimators=k,random_state=0)
    score=cross_val_score(classifier,X,y,cv=10)
    rfc_scores.append(score.mean())
plt.figure(figsize =(20,7))
plt.plot([k for k in range(1, 21)], rfc_scores, color = 'red')
for i in range(1,21):
    plt.text(i, rfc_scores[i-1], (i, round(rfc_scores[i-1],3)))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Estimators (K)')
plt.ylabel('Scores')
plt.title('XGBoost Classifier scores for different K values')
```

```python
classifier= XGBClassifier(n_estimators=20,random_state=0)
score=cross_val_score(classifier,X,y,cv=10)
print('% Accuracy :', round(score.mean()*100,4))
```

XGBoost Classifier was used  to predict the winners of the election. In order to know the optimum number of trees required to predict the result with highest accuracy, accuracy score for various values of k was plotted and k value was selected such that gives highest accuracy.

As the graph accuracy is maximum at k = 20, n_estimators=20 was selected.

An accuracy score of 95.3274 was achieved.
